```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import LabelEncoder

data1 = pd.DataFrame(data)

label_encoder = LabelEncoder()

data1['State'] = label_encoder.fit_transform(data1['State'])

print(data1)
```

```python
#import one hot encoder class from scikit-learn library
from sklearn.preprocessing import OneHotEncoder

#create a new dataframe
data2 = pd.DataFrame(data1)

#create an instance for one hot encoder
one_hot_encoder = OneHotEncoder()

#Encode the categorical column
encoded_data = one_hot_encoder.fit_transform(data1[['State']]).toarray()

#get the onehotencoded features out from the dataset2
encoded_feature_names = one_hot_encoder.get_feature_names_out(['State'])

#create the data2 with the encoded features
data2 = pd.DataFrame(encoded_data, columns = encoded_feature_names)

#drop the original country feature.
data1.drop(columns=['State'], inplace=True)

#concat the result of dataset2 and dataset3
data3 = pd.concat([data1, data2], axis=1)

#now print the result
data3
```

```python
data3.drop(columns=['State_0'], inplace=True)
```

```python
data3
```

```python
x=data3.iloc[:,[0,1,2,4,5]]
x
```

```python
y=data3.iloc[:,3]
y
```

```python
#import train_test_split class from the scikit-learn library model_selection
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
#import Linear Regression class from scikit-learn library
from sklearn.linear_model import LinearRegression

#create an instance
regressor = LinearRegression()

#fitting the training set
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
y_pred
```

```python
print("Train Score:", regressor.score(x_train, y_train))
print("Test Score:", regressor.score(x_test, y_test))
```
